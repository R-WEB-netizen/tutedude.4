2. Perform data preprocessing (handle missing data, normalize features if necessary).

```python
# Replace invalid zeros with NaN
columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[columns] = df[columns].replace(0, np.nan)
```

```python
# Fill missing values using median
imputer = SimpleImputer(strategy='median')
df[columns] = imputer.fit_transform(df[columns])
```

```python
# Check missing values after imputation
print("\nMissing values after preprocessing:")
print(df.isnull().sum())
```

```python
# Separate features and target
X = df.drop("Outcome", axis=1)
y = df["Outcome"]
```

```python
# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Display first five scaled rows
print("\nFirst 5 rows of scaled features:")
print(X_scaled[:5])
```

3. Split the dataset into training and testing sets.


```python
# Split the dataset into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,          # Features
    y,                 # Target variable
    test_size=0.20,    # 20% for testing
    random_state=42,   # Ensures reproducible results
    stratify=y         # Maintains class distribution
)
```

```python
print("Training Features Shape :", X_train.shape)
print("Testing Features Shape  :", X_test.shape)
print("Training Labels Shape   :", y_train.shape)
print("Testing Labels Shape    :", y_test.shape)
```

4. Implement the K-Nearest Neighbors (KNN) algorithm for prediction

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
print("predicted values:", y_pred)
```

```python
# Compare actual and predicted values
comparison = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

print("\nActual vs Predicted:")
print(comparison.head(10))
```

5. Evaluate the model using accuracy, confusion matrix, and other suitable metrics.




```python
# Accuracy
accuracy = np.mean(y_test == y_pred)
print("Accuracy of the KNN model diabetes dataset:", accuracy )


```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
#try different value of k
k_values = range(1, 21)
scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    score = knn.score(X_test, y_test)
    scores.append(score)

plt.plot(k_values, scores)
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('K value vs Accuracy')
plt.show()

best_k = k_values[scores.index(max(scores))]
print("Best k value:", best_k)
```
